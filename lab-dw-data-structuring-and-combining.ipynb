```python
# Your code goes here
import pandas as pd

url1='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url2='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

file0=pd.read_csv('labday2.csv')
df1=pd.read_csv(url1)
df2=pd.read_csv(url2)
```

```python
file1=df1.copy()
display(file1.info())
file1.head(2)
```

```python
file2=df2.copy()
display(file2.info())
file2=df2.copy()
file2.head(2)
```

```python
#df.rename(columns={"old_column_name1": "new_column_name1", "old_column_name2": "new_column_name2",...})
file1.rename(columns={'ST': 'State','ST': 'State'}, inplace=True)
file1.columns= file1.columns.map(lambda x:x.lower())
file2.columns= file2.columns.map(lambda x:x.lower())
print(file1.columns)
print(file2.columns)
```

```python
#organizing the tables so they have the same order, rearranging the column order for file 2 columns according to file 1
new_order = file1.columns

file2 = file2[new_order]
display(file1.head(1))
display(file2.head(1))
```

```python
df0=file0.copy()
df0.head(1)
```

```python
#now we concatenate the files 
df=pd.concat([file1, file2], axis=0).reset_index(drop=True)
#we check if the number of total rows is correct
display(df.info())
#we check how it look the labels
df.head(1)
```

```python
df.columns= df.columns.str.replace(' ','_')

#organizing the tables so they have the same order, rearranging the column order for file 2 columns according to file 1
new_or = df.columns
df0 = df0[new_or]
#we check if columns of both datasets have the same names
sorted(df.columns) == sorted(df0.columns)
```

```python
#before concatenating, we need to clean the df data as the df0 (from previous lab) is already clean.
# Gender standarization
#gender={'Femal':'F', 'Male':'M', 'female':'F'}
#df.gender=df.gender.map(gender)
df.gender.unique()
```

```python
#we fix the naming to aling values
df.gender=df.gender.map(lambda x:
                'F'if isinstance(x,str) and x[0].lower()=='f' else
                'M' if isinstance(x,str) and x[0].lower()=='m'else
                x)
#we check if columns of both datasets have the same unique values for the column
#for now without considering NaN
sorted(df.gender.dropna().unique()) == sorted(df0.gender.dropna().unique())
```

```python
df.state.unique()
```

```python
#we fix the naming to aling values
df.state=df.state.replace('AZ','Arizona')
#we check if columns of both datasets have the same unique values for the column
#for now without considering NaN
sorted(df.state.dropna().unique()) == sorted(df0.state.dropna().unique())
```

```python
df.education.unique()
```

```python
#we fix the naming to aling values
df.education=df.education.replace('Bachelors','Bachelor')
#we check if columns of both datasets have the same unique values for the column
#for now without considering NaN
sorted(df.education.dropna().unique()) == sorted(df0.education.dropna().unique())
```

```python
df.customer_lifetime_value.unique()
```

```python
#In Customer Lifetime Value, delete the % character
df.customer_lifetime_value=df.customer_lifetime_value.dropna().str.replace('%','',regex=False)
```

```python
df.vehicle_class.unique()
```

```python
df.vehicle_class= df.vehicle_class.replace({
    "Sports Car": 'Luxury',
    "Luxury SUV": 'Luxury',
    "Luxury Car": 'Luxury'})
#we check if columns of both datasets have the same unique values for the column
#for now without considering NaN
sorted(df.vehicle_class.dropna().unique()) == sorted(df0.vehicle_class.dropna().unique())
```

```python
#we check which columns have different types:
df.apply(lambda col: col.dtype)==df0.apply(lambda col: col.dtype)
```

```python
#let's check for null
df.isnull()
#for the rows with nulls in all columns we drop
df[df.isnull()].all(axis=1)
df=df.dropna(how='all')
```

```python
#let see if it worked: and yes it did.
df.isnull().sum()
```

```python
#let's fill gender column with the mode
df.gender=df.gender.fillna(df.gender.mode()[0])
df.gender.isnull().sum()
```

```python
#df[df.customer_lifetime_value.isnull()]
display(df.policy_type.unique())
life_df= df[df.customer_lifetime_value.isnull()==False]
life_df.customer_lifetime_value=life_df.customer_lifetime_value.astype(float)
personal_mode=round(life_df[life_df.policy_type=='Personal Auto'].customer_lifetime_value.mean())
special_mode=round(life_df[life_df.policy_type=='Special Auto'].customer_lifetime_value.mean())
corp_mode=round(life_df[life_df.policy_type=='Corporate Auto'].customer_lifetime_value.mean())

#filling the column customer value
fill_map={'Personal Auto': personal_mode,'Special Auto': special_mode,
          'Corporate Auto': corp_mode}

for policy,value in fill_map.items():
    mask=df.policy_type==policy
    df.loc[mask,'customer_lifetime_value']= df.loc[mask,'customer_lifetime_value'].fillna(value)

display(df.customer_lifetime_value.isnull().sum())
```

```python
df.info()
```

```python
# Customer lifetime value should be integer to match the other dataframe

df.customer_lifetime_value=round(df.customer_lifetime_value.astype(float))
df.customer_lifetime_value=df.customer_lifetime_value.astype(int)
df.customer_lifetime_value.dtype
```

```python
#file0 which is the same df0 now, it was cleaned in the lab before but let's check if is clean with nulls
#let's check for null
display(df0.isnull().sum())
#as this was the last cleaning step from the previous lab, it is inferred the dataset is clean now.
```

```python
# Next column dtype change, should be integer to match the other dataframe
df.total_claim_amount=df.total_claim_amount.astype(int)
df.total_claim_amount.dtype
```

```python
# Number of open complaints should be integer to match the other dataframe
df.number_of_open_complaints
```

```python
#take the middle number only as previous lab cleaning file0 or df0
df.number_of_open_complaints=df.number_of_open_complaints.str.split('/').map(lambda x: x[1] if isinstance(x,list) else x )
df.number_of_open_complaints
```

```python
display(df.number_of_open_complaints.isnull().sum())
#we fill with mean
mean_comp=round(df[df.number_of_open_complaints.isnull()==False].number_of_open_complaints.astype(int).mean())
df.number_of_open_complaints=df.number_of_open_complaints.fillna(mean_comp)
#check for null values
display(df.number_of_open_complaints.isnull().sum())
```

```python
#now that there are no more nan we convert to int
df.number_of_open_complaints=df.number_of_open_complaints.astype(int)
df.number_of_open_complaints.dtype
```

```python
#we check now if all columns have same dtype
df.apply(lambda col: col.dtype)==df0.apply(lambda col: col.dtype)
```

```python
#now we concatenate
# Concatenate the sales, and sales_2 vertically (along rows)
pd.concat([df, df0], axis=0).reset_index(drop=True)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
#Aim: which sales channel brought the most sales in terms of total revenue
```

```python
sales=datf.copy()
display(sales.info())
sales.head(2)
```

```python
#Checking data values, organisation and null values
sales['unnamed:_0'].unique() # this seems like indexes
```

```python
#Using pivot, create a summary table showing the total revenue for each sales channel
#(branch, call center, web, and mail).
```

```python
#we will calculate revenue as:
sales['revenue'] = round(sales.customer_lifetime_value -sales.total_claim_amount)
sales.head(1)
```

```python
sales_piv = sales.pivot_table(
    index="sales_channel", 
    values="revenue",
    aggfunc="sum",
)
display(sales_piv)
```

```python
#Analyze the resulting table to draw insights
print(f"the branch that has the highest revenue is the Agent channel with a revenue of {sales_piv.revenue[0]}")
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
sales_muliindex=['gender','education']

custogender_piv = sales.pivot_table(
    index=sales_muliindex, 
    values="customer_lifetime_value",
    aggfunc="mean",
)
display(custogender_piv)
```

```python
#from the table above, one can infer that among females with High School or Below 
#are actually the ones bringing more customer value.
#for the case of men, is the opposite, the men with the hight level of education
#like Master are the ones bringing more customer lifetime value.
```
